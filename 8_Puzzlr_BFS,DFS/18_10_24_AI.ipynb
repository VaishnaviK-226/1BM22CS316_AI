```python
def vacuum_world():
    # Initializing goal_state
    goal_state = {'A': '0', 'B': '0'}  # 0 indicates Clean, 1 indicates Dirty
    cost = 0

    location_input = input("Enter Location of Vacuum (A/B): ").strip()
    status_input = input(f"Enter status of {location_input} (0 for Clean, 1 for Dirty): ").strip()
    status_input_complement = input("Enter status of other room (0 for Clean, 1 for Dirty): ").strip()

    print("Initial Location Condition: " + str(goal_state))

    if location_input == 'A':
        print("Vacuum is placed in Location A.")

        if status_input == '1':  # Location A is Dirty
            print("Location A is Dirty.")
            goal_state['A'] = '0'  # Clean A
            cost += 1  # Cost for cleaning A
            print("Cost for CLEANING A: " + str(cost))
            print("Location A has been Cleaned.")

        if status_input_complement == '1':  # If B is Dirty
            print("Location B is Dirty.")
            print("Moving right to Location B.")
            cost += 1  # Cost for moving right
            print("COST for moving RIGHT: " + str(cost))
            goal_state['B'] = '0'  # Clean B
            cost += 1  # Cost for cleaning B
            print("COST for SUCK: " + str(cost))
            print("Location B has been Cleaned.")

    elif location_input == 'B':
        print("Vacuum is placed in Location B.")

        if status_input == '1':  # Location B is Dirty
            print("Location B is Dirty.")
            goal_state['B'] = '0'  # Clean B
            cost += 1  # Cost for cleaning B
            print("COST for CLEANING B: " + str(cost))
            print("Location B has been Cleaned.")

        if status_input_complement == '1':  # If A is Dirty
            print("Location A is Dirty.")
            print("Moving left to Location A.")
            cost += 1  # Cost for moving left
            print("COST for moving LEFT: " + str(cost))
            goal_state['A'] = '0'  # Clean A
            cost += 1  # Cost for cleaning A
            print("COST for SUCK: " + str(cost))
            print("Location A has been Cleaned.")

    else:
        print("Invalid location entered.")

    # Done cleaning
    print("GOAL STATE: ")
    print(goal_state)
    print("Performance Measurement: " + str(cost))

# To run the function
vacuum_world()
```

```python
from collections import deque

class PuzzleState:
    def __init__(self, board, empty_tile_pos, moves=0):
        self.board = board  # Current configuration of the puzzle
        self.empty_tile_pos = empty_tile_pos  # Position of the blank space
        self.moves = moves  # Number of moves taken to reach this state

    def is_goal(self):
        # Check if the current board matches the goal configuration
        return self.board == [1, 2, 3, 4, 5, 6, 7, 8, 0]

    def get_possible_moves(self):
        moves = []
        x, y = self.empty_tile_pos
        directions = {
            'UP': (x - 1, y),
            'DOWN': (x + 1, y),
            'LEFT': (x, y - 1),
            'RIGHT': (x, y + 1)
        }

        for direction, (new_x, new_y) in directions.items():
            if 0 <= new_x < 3 and 0 <= new_y < 3:  # Check bounds
                new_board = self.board[:]  # Create a new board configuration
                # Swap the empty tile with the adjacent tile
                new_board[x * 3 + y], new_board[new_x * 3 + new_y] = new_board[new_x * 3 + new_y], new_board[x * 3 + y]
                moves.append((new_board, (new_x, new_y), direction))  # Include direction for tracing
        return moves

def dfs(initial_state):
    stack = [initial_state]
    visited = set()

    while stack:
        current_state = stack.pop()
        visited.add(tuple(current_state.board))  # Mark this state as visited

        if current_state.is_goal():  # Goal test
            return current_state.moves

        for new_board, new_empty_pos, action in current_state.get_possible_moves():
            if tuple(new_board) not in visited:
                stack.append(PuzzleState(new_board, new_empty_pos, current_state.moves + 1))

    return -1  # No solution found

def bfs(initial_state):
    queue = deque([initial_state])
    visited = set()

    while queue:
        current_state = queue.popleft()
        visited.add(tuple(current_state.board))  # Mark this state as visited

        if current_state.is_goal():  # Goal test
            return current_state.moves

        for new_board, new_empty_pos, action in current_state.get_possible_moves():
            if tuple(new_board) not in visited:
                queue.append(PuzzleState(new_board, new_empty_pos, current_state.moves + 1))

    return -1  # No solution found

# Example usage
initial_board = [1, 2, 3, 4, 5, 0, 7, 8, 6]  # Example initial state
initial_empty_pos = (1, 2)  # Position of the blank tile
initial_state = PuzzleState(initial_board, initial_empty_pos)

print("DFS moves:", dfs(initial_state))
print("BFS moves:", bfs(initial_state))
```
